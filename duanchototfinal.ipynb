```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import requests
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.select import Select
from selenium.webdriver.edge.service import service
from bs4 import BeautifulSoup
driver = webdriver.Edge(executable_path='C:\\Program Files (x86)\\msedgedriver.exe')
#toi trang
with open('dataxechotot.csv','w',encoding='utf-8') as file:
    file.write( 'Hãng'+ ';' + 'Dòng xe' +';' + 'Giá' + ';' + 'Số Chỗ'
                        + ';' + 'số Km đã chạy' + ';' +'Tình Trạng' + ';' +'Năm sản Xuất' + ';' +
                       'Hộp số' + ';' +'Nhiên liệu' + ';' +'Xuất sứ' + ';' +'Kiểu Dáng' + ';' +'Links'+'\n')

for k in range(1,1432):
    driver.get('https://xe.chotot.com/mua-ban-oto?page='+str(k))
    driver.implicitly_wait(1)
    #lay link
    #################################################################################
    page_source = BeautifulSoup(driver.page_source)
    profiles = page_source.find_all('a', class_ = 'AdItem_adItem__2O28x')
    allprofile=[]
    for profile in profiles:
        profile_ID = profile.get('href')
        profile_URL = "https://xe.chotot.com" + profile_ID
        allprofile.append(profile_URL)
    #################################################################################
    tenxe=driver.find_elements(By.XPATH,"//h3[contains(@class,'commonStyle_adTitle__1VYG8')]")
    hang = []
    dongxe= []
    gia= []
    tinhtrang= []
    hopso= []
    nhienlieu= []
    xuatsu= []
    loaixe= []
    namsanxuat = []
    socho = []
    sokm=[]
    #################################################################################
    for p in range(len(tenxe)):
        page = requests.get(allprofile[p])
        soup1=BeautifulSoup(page.content,'html.parser')
        #lay content o day
        try:
            hang2=soup1.find(itemprop='carbrand').get_text()
            hang.append(hang2)
        except AttributeError:
            hang.append('null')
###################################################################################################################

###################################################################################################################
        try:
            hang2=soup1.find(itemprop='carbrand').get_text()
            hang.append(hang2)
        except AttributeError:
            hang.append('null')
        
        try:
            gia2= soup1.find(itemprop='price').get_text()
            gia.append(gia2)
        except AttributeError:
            gia.append('null')
        try:
            dongxe2= soup1.find(itemprop='carmodel').get_text()
            dongxe.append(dongxe2)
        except AttributeError:
            dongxe.append('0')
        try:
            sokm2= soup1.find(itemprop='mileage_v2').get_text()
            sokm.append(sokm2)
        except AttributeError:
            sokm.append('0')
        try:
            tinhtrang2= soup1.find(itemprop='condition_ad').get_text()
            tinhtrang.append(tinhtrang2)
        except AttributeError:
            tinhtrang.append('null')
        try:
            namsanxuat2= soup1.find(itemprop='mfdate').get_text()
            namsanxuat.append(namsanxuat2)
        except AttributeError:
            namsanxuat.append('null')
        try:    
            hopso2= soup1.find(itemprop='gearbox').get_text()
            hopso.append(hopso2)
        except AttributeError:
            hopso.append('null')
        try:
            nhienlieu2= soup1.find(itemprop='fuel').get_text()
            nhienlieu.append(nhienlieu2)
        except AttributeError:
            nhienlieu.append('null')
        try:
            xuatsu2= soup1.find(itemprop='carorigin').get_text()
            xuatsu.append(xuatsu2)
        except AttributeError:
            xuatsu.append('null')
        try:
            loaixe2= soup1.find(itemprop='cartype').get_text()
            loaixe.append(loaixe2)
        except AttributeError:
            loaixe.append('null')
        try:
            socho2= soup1.find(itemprop='carseats').get_text()
            socho.append(socho2)
        except AttributeError:
            socho.append('null')       
    with open('dataxechotot.csv','a',encoding='utf-8') as file:
        for i in range(len(tenxe)):
            file.write( hang[i]+ ';' + dongxe[i] +';'+  gia[i] + ';' + socho[i]
                        + ';' +sokm[i] + ';' +tinhtrang[i] + ';' +namsanxuat[i] + ';' +
                       hopso[i] + ';' +nhienlieu[i] + ';' +xuatsu[i] + ';' +loaixe[i]  + ';' +  allprofile[i] +'\n')
    print('Hoan thanh trang'+ ' ' + str(k))
    #################################################################################
driver.quit()
print ('xong rồi bé ơi ')
```
