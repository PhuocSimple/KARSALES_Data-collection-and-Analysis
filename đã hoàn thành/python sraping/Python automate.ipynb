```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import requests
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.select import Select
from selenium.webdriver.chrome.service import Service
from bs4 import BeautifulSoup
import pandas as pd
import datetime 
s=Service("C://Program Files (x86)//chromedriver.exe")

#toi trang
user_agent = "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/98.0.4758.82 Safari/537.36"

options = webdriver.ChromeOptions()
options.headless = True
options.add_argument(f'user-agent={user_agent}')
options.add_argument("--window-size=1920,1080")
options.add_argument('--ignore-certificate-errors')
options.add_argument('--allow-running-insecure-content')
options.add_argument("--disable-extensions")
options.add_argument("--proxy-server='direct://'")
options.add_argument("--proxy-bypass-list=*")
options.add_argument("--start-maximized")
options.add_argument('--disable-gpu')
options.add_argument('--disable-dev-shm-usage')
options.add_argument('--no-sandbox')
driver = webdriver.Chrome(service=s, options=options)
hang = []
dongxe= []
gia= []
tinhtrang= []
hopso= []
nhienlieu= []
xuatsu= []
loaixe= []
namsanxuat = []
socho = []
sokm=[]
allprofile=[]
now = datetime.datetime.now()
now = now.strftime("%d-%m-%y %H:%M:%S")
now1 = []
now1.append(now)
name = int(input('Nhập số trang bạn muốn'))+1
for k in range(1,name):
    driver.get('https://xe.chotot.com/mua-ban-oto?page='+str(k))
    
    driver.implicitly_wait(1)
    #lay link
    #################################################################################
    
    page_source = BeautifulSoup(driver.page_source,features="html.parser")
    profiles = page_source.find_all('a', class_ = 'AdItem_adItem__2O28x')
    for profile in profiles:
        profile_ID = profile.get('href')
        profile_URL = "https://xe.chotot.com" + profile_ID
        allprofile.append(profile_URL)
    #################################################################################
    tenxe=driver.find_elements(By.XPATH,"//h3[contains(@class,'commonStyle_adTitle__1VYG8')]")
    #################################################################################
    for p in range(len(tenxe)):
        page = requests.get(allprofile[p])
        soup1=BeautifulSoup(page.content,'html.parser')
        #lay content o day
###################################################################################################################
        try:
            hang2=soup1.find(itemprop='carbrand').get_text()
            hang.append(hang2)
        except AttributeError:
            hang.append('null')
        try:
            gia2= soup1.find(itemprop='price').get_text()
            gia.append(gia2)
        except AttributeError:
            gia.append('null')
        try:
            dongxe2= soup1.find(itemprop='carmodel').get_text()
            dongxe.append(dongxe2)
        except AttributeError:
            dongxe.append('0')
        try:
            sokm2= soup1.find(itemprop='mileage_v2').get_text()
            sokm.append(sokm2)
        except AttributeError:
            sokm.append('0')
        try:
            tinhtrang2= soup1.find(itemprop='condition_ad').get_text()
            tinhtrang.append(tinhtrang2)
        except AttributeError:
            tinhtrang.append('null')
        try:
            namsanxuat2= soup1.find(itemprop='mfdate').get_text()
            namsanxuat.append(namsanxuat2)
        except AttributeError:
            namsanxuat.append('null')
        try:    
            hopso2= soup1.find(itemprop='gearbox').get_text()
            hopso.append(hopso2)
        except AttributeError:
            hopso.append('null')
        try:
            nhienlieu2= soup1.find(itemprop='fuel').get_text()
            nhienlieu.append(nhienlieu2)
        except AttributeError:
            nhienlieu.append('null')
        try:
            xuatsu2= soup1.find(itemprop='carorigin').get_text()
            xuatsu.append(xuatsu2)
        except AttributeError:
            xuatsu.append('null')
        try:
            loaixe2= soup1.find(itemprop='cartype').get_text()
            loaixe.append(loaixe2)
        except AttributeError:
            loaixe.append('null')
        try:
            socho2= soup1.find(itemprop='carseats').get_text()
            socho.append(socho2)
        except AttributeError:
            socho.append('null')

    print('Hoan thanh trang'+ ' ' + str(k)+'/' +str(name - 1 ))
df1=pd.DataFrame(list(zip(hang, dongxe,gia,socho,sokm,tinhtrang,namsanxuat,hopso,nhienlieu,xuatsu,loaixe,allprofile)),columns =['Hãng', 'Dòng xe','Giá','Số Chỗ','số Km đã chạy','Tình Trạng','Năm sản Xuất','Hộp số','Nhiên liệu','Xuất sứ','Kiểu Dáng','Links'])
df1['Thời gian cập nhập'] = now
df1.to_csv(r'C:\Users\Phuoc Phe Phon\Desktop\pYTHON\dataxechotot2.csv',index=False)
    #################################################################################
driver.quit()
print ('xong rồi bé ơi ')
```
